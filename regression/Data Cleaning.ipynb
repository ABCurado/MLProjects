## Data Cleaning

In this Workbook the findings of the EDA are taken and the data is wrangled accordingly.

```python
import pandas as pd
```

```python
df = pd.read_pickle("./raw.pkl")
```

```python
data = pd.DataFrame()
```

```python
selected_features = ['Ageing index',
                     'Population density',
                     'migrants_Males',
                     'labour productivity',
                     "x",
                     "y",
                    "Unemployment_Rate",
                    "unemp_45 - 54",
                     "higher_female",
                     "no_male",
                     "avg_wage",
                     "Capital revenue",
                     'PPD/PSD'
                    ]
```

```python
for i in selected_features:
    data[i] = df[i]
```

```python
# impute avg_wage with mean
data.loc[df["avg_wage"]==0,"avg_wage"] = data.loc[df["avg_wage"]>0,"avg_wage"].mean()
```

```python
df.loc[df["Capital revenue"]>1500,"Capital revenue"]
```

```python
# impute Corvo
data.loc[data["Capital revenue"]>1500,"Capital revenue"] = 1571.31
```

```python
# dump to pickle
data.to_pickle("./unemp.pkl")
```
