```python
import pandas as pd
import numpy as np
from tensorflow import keras
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from keras import models
from keras import layers
import datetime
from datetime import timedelta
```

```python
df = pd.read_excel("ml_project1_data.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
datetimeFormat = '%Y-%m-%d'
today = "2019-04-12"
days_customer = []

for i in df["Dt_Customer"]:
    
    diff = datetime.datetime.strptime(today, datetimeFormat) - datetime.datetime.strptime(i, datetimeFormat)
    days_customer.append(diff.days)

df["DaysCustomer"] = days_customer
```

```python
df = df.drop(["ID"], axis =1)
df = df.drop(["Z_Revenue"], axis = 1)
df = df.drop(["Z_CostContact"], axis = 1)
df = df.drop(["Dt_Customer"], axis = 1)
```

```python
df.head()
```

```python
df["Education"].unique()
```

```python
df["Marital_Status"].unique()
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
training_set, test_set, y_training_set, y_test_set = train_test_split(df.drop(["Response"], axis=1), df["Response"], test_size=0.2, random_state=42)
```

```python
training_set.info()
```

```python
# Feature Scaling
scaler = MinMaxScaler()
training_set[["Year_Birth","Income","Kidhome","Teenhome","Recency","MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds","NumDealsPurchases","NumCatalogPurchases","NumStorePurchases","NumWebVisitsMonth","days_customer"]] = scaler.fit_transform(training_set[["Year_Birth","Income","Kidhome","Teenhome","Recency","MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds","NumDealsPurchases","NumCatalogPurchases","NumStorePurchases","NumWebVisitsMonth","days_customer"]])
test_set[["Year_Birth","Income","Kidhome","Teenhome","Recency","MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds","NumDealsPurchases","NumCatalogPurchases","NumStorePurchases","NumWebVisitsMonth","days_customer"]] = scaler.transform(test_set[["Year_Birth","Income","Kidhome","Teenhome","Recency","MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds","NumDealsPurchases","NumCatalogPurchases","NumStorePurchases","NumWebVisitsMonth","days_customer"]])
```

```python
training_set.head()
```

```python
training_set.shape
```

```python
model = models.Sequential()
model.add(layers.Dense(12, activation="relu", input_dim=36))
model.add(layers.Dense(12, activation="relu"))
model.add(layers.Dense(1, activation="sigmoid"))
model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(training_set, y_training_set, epochs=100)
y_predicted = model.predict(training_set)
```

```python
model.evaluate(test_set, y_test_set)
```
