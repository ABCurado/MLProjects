```python
def missing_values_reporter(df):    
    na_count = df.isna().sum() 
    ser = na_count[na_count > 0]
    return pd.DataFrame({"N missings": ser, "% missings": ser.divide(df.shape[0])})

df_miss=missing_values_reporter(df)
df_miss
```

```python
import matplotlib
import matplotlib.pyplot as plt

import numpy as np
import seaborn as sns

cutoff_ = 0.03
cutoff_list = [cutoff_, cutoff_]
ax = df_miss.sort_values('% missings', ascending=False).plot.bar(y="% missings", 
                                                                 color="gray", 
                                                                 title="{}% cutoff line on missing values".format(cutoff_*100), legend=False)
ax.set_xlabel("Categorical features with missing values", size=12)
ax.set_ylabel("Proportion of missings")
ax.plot([-1, len(df_miss.index)], cutoff_list,'r--', lw=2)
ax.set_xticklabels(df_miss.index, rotation=60, size=9)
plt.show()
```

```python
def describe_cat(df, list_cfeatures, target):
    cat_list = []
    for feature in df[list_cfeatures]:    
        cat_list.append(df.groupby([feature]).agg({target: ['count', "mean"]}))
    return pd.concat(cat_list, axis=0, keys=list_cfeatures)
```

```python
def categorical_discrimination_plot(df, list_cfeatures, target):
    df_cat = describe_cat(df, list_cfeatures, target)

    cutoff_ = df[target].mean()
    cutoff_list = [cutoff_, cutoff_]
    
    for feature in list_cfeatures:
        df_cat1=df_cat.iloc[:, 1][feature].sort_values(ascending=False)
        ax = df_cat1.plot.bar(color="gray", title="{}% cutoff line on discrimination ability".format(round(cutoff_*100, 2)), legend=False)
        ax.set_xticklabels(df_cat1.index, rotation=60, size=9)      
        ax.set_ylabel("Proportion of respondents")
        ax.set_xlabel(feature)
        ax.plot([-1, len(df_cat)], cutoff_list,'r--', lw=2)
        plt.show()
```

```python
categorical_discrimination_plot(df, df.select_dtypes(include=["object", "category"]).columns, "Response")
```

```python
def categorical_discrimination_bivariate_plot(df, list_cfeatures, target):
    target = "Response"
    feature = 'Education'
    
    for feature in list_cfeatures:
        # df: normal
        df_biv = df.groupby([feature, target])[feature].count().unstack(target).fillna(0)

        # plot normal 
        fig, axes = plt.subplots(nrows=1, ncols=2)
        df_biv[[0, 1]].plot(kind='bar', stacked=True, ax=axes[0])
        axes[0].set(title="Stacked bar plot (freq)")
        axes[0].set_xticklabels(df_biv.index, rotation=60, size=8)  

        # df: %
        df_biv_rel=df_biv.div(df_biv[0]+df_biv[1], axis=0)*100
        df_biv_rel[[0, 1]].plot(kind='bar', stacked=True, ax=axes[1])
        axes[1].set(title="Stacked bar plot (%)")
        axes[1].set_xticklabels(df_biv.index, rotation=60, size=8)  

        # plot %
        axes[0].set(title="Bivariate bar (stacked)")
        plt.show()
```

```python
categorical_discrimination_bivariate_plot(df, df.select_dtypes(include=["object", "category"]).columns, "Response")
```

### Scatter

```python
from matplotlib.gridspec import GridSpec

def personalized_scatter(df, x_f, y_f, target, n_bins=20): 
    colors = ("gray", "red")
    groups = ("non-respondents", "respondents")

    # subset
    X_0 = df.loc[df[target]==0, [x_f, y_f]].values
    X_1 = df.loc[df[target]==1, [x_f, y_f]].values    
    data = ((X_0[:, 0], X_0[:, 1]), (X_1[:, 0], X_1[:, 1])) 

    fig = plt.figure(figsize=(6,6))
    gs = GridSpec(4,4)
    ax_joint = fig.add_subplot(gs[1:4,0:3])
    ax_marg_x = fig.add_subplot(gs[0,0:3])
    ax_marg_y = fig.add_subplot(gs[1:4,3])

    for data, color, group in zip(data, colors, groups):
        x, y = data
        ax_joint.scatter(x, y, alpha=.6, c=color, edgecolors='none', s=30, label=group)
         
    ax_marg_x.hist(df[x_f].values, color="lightgray")
    ax_marg_y.hist(df[y_f].values, orientation="horizontal", color="lightgray")

    ax_xdens = ax_marg_x.twinx()
    ax_xdens.hist(X_0[:, 0], bins=n_bins, histtype='step', color='black', density=True, align='mid')
    ax_xdens.hist(X_1[:, 0], bins=n_bins, histtype='step', color='r', density=True, align='mid')
    ax_xdens.axis('off')

    ax_ydens = ax_marg_y.twiny()
    ax_ydens.hist(X_0[:, 1], bins=n_bins, histtype='step', color='black', density=True, align='mid',orientation='horizontal')
    ax_ydens.hist(X_1[:, 1], bins=n_bins, histtype='step', color='r', density=True, align='mid',orientation='horizontal')
    ax_ydens.axis('off')

    # turn off tick labels on marginals
    plt.setp(ax_marg_x.get_xticklabels(), visible=False)
    plt.setp(ax_marg_y.get_yticklabels(), visible=False)

    # set labels on joint
    ax_joint.set_xlabel(x_f)
    ax_joint.set_ylabel(y_f)

    # set labels on marginals
    ax_joint.legend()
    plt.show()
```

```python
for i in df:
    if df[i].dtype == "float64" or df[i].dtype == "int64":
        
        personalized_scatter(df, i,"MntGoldProds", "Response")
```
